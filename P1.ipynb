```python
thetaobs=linspace(0,2*pi,2000)
def my_model(theta,r0,r1,m,phi,x0,y0):
    test=zeros((ly,lx))
    r=r0+r1*sin(m*theta+phi)
    x=(r*cos(theta)+x0).astype(int)
    y=(r*sin(theta)+y0).astype(int)
    #if((x>=0 and x<=lx)and(y>=0 and y<=ly)):
    a=x[(x>=0) & (x<=lx) & (y>=0) & (y<=ly)]
    b=y[(x>=0) & (x<=lx) & (y>=0) & (y<=ly)]
    test[b,a]=1
    return test
```

```python
def likelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum((y_obs-y_model)**2)
    return (-chi_squared)
```

```python
r0_walk = empty((0)) #this is an empty list to keep all the steps
r1_walk = empty((0))
m_walk = empty((0))
phi_walk = empty((0))
x0_walk = empty((0))
y0_walk = empty((0))
l_walk = empty((0))

r0_walk = append(r0_walk, 80)
r1_walk = append(r1_walk, random.random())
m_walk = append(m_walk, random.random())
phi_walk = append(phi_walk, random.random())
x0_walk = append(x0_walk, lx/2)
y0_walk = append(y0_walk, ly/2)

y_init = my_model(thetaobs, r0_walk[0], r1_walk[0],m_walk[0], phi_walk[0], x0_walk[0],y0_walk[0])
l_walk = append(l_walk, likelihood(imgnorm, y_init))
print r0_walk
print r1_walk
print m_walk
print phi_walk
print x0_walk
print y0_walk
print l_walk
```

```python
n_iterations = 20000 #this is the number of iterations I want to make
for i in range(n_iterations):
    r0_prime = np.random.normal(r0_walk[i], 0.1) 
    r1_prime = np.random.normal(r1_walk[i], 0.1)
    m_prime = np.random.normal(m_walk[i], 0.1)
    phi_prime = np.random.normal(phi_walk[i], 0.1) 
    x0_prime = np.random.normal(x0_walk[i], 0.1)
    y0_prime = np.random.normal(y0_walk[i], 0.1)

    y_init = my_model(thetaobs, r0_walk[i], r1_walk[i], m_walk[i], phi_walk[i], x0_walk[i], y0_walk[i])
    y_prime = my_model(thetaobs, r0_prime, r1_prime, m_prime, phi_prime, x0_prime, y0_prime)
    
    l_prime = likelihood(imgnorm, y_prime)
    l_init = likelihood(imgnorm, y_init)
    
    alpha = exp(l_prime-l_init)
    if(alpha>=1.0):
        r0_walk  = append(r0_walk,r0_prime)
        r1_walk  = append(r1_walk,r1_prime)
        m_walk  = append(m_walk,m_prime)
        phi_walk  = append(phi_walk,phi_prime)
        x0_walk  = append(x0_walk,x0_prime)
        y0_walk  = append(y0_walk,y0_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta<=alpha):
            r0_walk  = append(r0_walk,r0_prime)
            r1_walk  = append(r1_walk,r1_prime)
            m_walk  = append(m_walk,m_prime)
            phi_walk  = append(phi_walk,phi_prime)
            x0_walk  = append(x0_walk,x0_prime)
            y0_walk  = append(y0_walk,y0_prime)
            l_walk = append(l_walk, l_prime)
        else:
            r0_walk = append(r0_walk,r0_walk[i])
            r1_walk = append(r1_walk,r1_walk[i])
            m_walk = append(m_walk,m_walk[i])
            phi_walk = append(phi_walk,phi_walk[i])
            x0_walk = append(x0_walk,x0_walk[i])
            y0_walk = append(y0_walk,y0_walk[i])
            l_walk = append(l_walk, l_init)
```
